```python
import numpy as np
import pandas as pd
```

```python
df_2022 = pd.read_csv('C:/Users/h_olv/Downloads/rainfall_2022.csv', encoding='cp949')
df_2021 = pd.read_csv('C:/Users/h_olv/Downloads/rainfall_2021.csv', encoding='cp949')
df_2020 = pd.read_csv('C:/Users/h_olv/Downloads/rainfall_2020.csv', encoding='cp949')
df_2019 = pd.read_csv('C:/Users/h_olv/Downloads/rainfall_2019.csv', encoding='cp949')
df_2018 = pd.read_csv('C:/Users/h_olv/Downloads/rainfall_2018.csv', encoding='cp949')
```

```python
print(len(df_2022))
print(len(df_2021))
print(len(df_2020))
print(len(df_2019))
print(len(df_2018))
```

```python
df = pd.concat([df_2022, df_2021, df_2020, df_2019, df_2018], ignore_index=True)
df
```

```python
df['지점명'].unique()
```

종로구와 동작구가 없음

'기상청', '한강', '남현', '현충원', '관악(레)' drop

```python
df = df[~df['지점명'].isin(['기상청', '한강', '남현', '현충원', '관악(레)'])]
df
```

```python
df['날짜'] = pd.to_datetime(df['일시']).dt.date
df
```

```python
df['cum_rainfall'] = df.groupby(['지점명', '날짜'])['강수량(mm)'].cumsum()
df
```

```python
df_3mm = df[df['강수량(mm)'] >= 3]
df_3mm
```
